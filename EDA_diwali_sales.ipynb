```python
import plotly.express as px
```

# **1. How many males and females plot the count of gender**

```python
fig = px.histogram(diwali, x='Gender',color='Gender', color_discrete_sequence=["red","blue"],width=600,height=500)
fig.show()
```

# **2. for male and female**
#**Give me the sum of amt wrt each gender and plot the graph**


```python
diwlali = diwali.groupby('Gender')['Amount'].sum().reset_index()
fig = px.histogram(diwali, x='Gender', y='Amount',
                   title='Total Amount by Gender',
                   color='Gender', color_discrete_sequence=["blue","yellow"],width=600,height=600)
fig.show()
```

# **3. count of each age group with individual count of age group wrt gender**

```python
diwali.groupby(['Age Group', 'Gender'])['Gender'].count()
fig = px.bar(diwali, x='Age Group', y='Gender', title='count of age group wrt gender',
                   color='Age Group')
fig.show()
```

# **4. plot the total amt spent by each age group**

```python
diwali = diwali.groupby('Age Group')['Amount'].sum().reset_index()
fig = px.histogram(diwali, x='Age Group', y='Amount', title='Total Amount by Age Group',color='Age Group',
                   color_discrete_sequence=["violet","blue","red","yellow","green","black","orange"])
fig.show()
```

# **5. Total amt of top 10 states**

```python
diwali = diwali.groupby('State')['Amount'].sum().reset_index().sort_values(by='Amount', ascending=False).head(10)
fig = px.histogram(diwali, x='State', y='Amount', title='Total Amount by State (Top 10)',color='State',
                   color_discrete_sequence=["violet","blue","red","yellow","green","black","orange","pink","purple","brown"])
fig.show()
```

# **6. plot the graph that depicts total number of orders from top 10 states**

```python
diwali = diwali.groupby('State')['Orders'].count().reset_index().sort_values(by='Orders', ascending=False).head(10)
fig = px.histogram(diwali, x='State', y='Orders', title='Total Orders by State (Top 10)',
             color='State',color_discrete_sequence=["brown","purple","red","yellow","orange","black","green","pink","blue","violet"])
fig.show()
```

# **7. plot me a graph that shows comparison between no of married and unmarried people**

```python
diwali_counts = diwali['Marital_Status'].value_counts().reset_index()
diwali_counts.columns = ['Marital_Status','Count']
fig = px.histogram(diwali_counts, x='Marital_Status', y='Count', title='Number of Married vs. Unmarried People',
                   color='Marital_Status',color_discrete_sequence=["green","yellow"],width=600)
fig.show()
```

# **8. plot me amt spent by each male and each femlae wrt to marital status**

```python
diwali = diwali.groupby(['Gender', 'Marital_Status'])['Amount'].sum().reset_index()
fig = px.bar(diwali, x='Gender', y='Amount', color='Marital_Status',
             title='Amount Spent by Gender and Marital Status', barmode='group',width=800,height=600)
fig.show()
```

# **9. plot me the count of each occupation that is present in dataset**

```python

diwali_occupations = diwali['Occupation'].value_counts().reset_index()
diwali_occupations.columns = ['Occupation', 'Count']
fig = px.histogram(diwali_occupations, x='Occupation', y='Count', title='Count of Each Occupation',
                   color='Occupation',color_discrete_sequence=["violet","blue","red","yellow","green","black","orange","pink","purple","brown"])
fig.show()
```

# **10. Plot the amt spent by each occupation in descending order**

```python
diwali = diwali.groupby('Occupation')['Amount'].sum().reset_index().sort_values(by='Amount', ascending=False)
import plotly.express as px
fig = px.histogram(diwali, x='Occupation', y='Amount', title='Amount Spent by Occupation',
                   color='Occupation',color_discrete_sequence=["violet","blue","red","yellow","green","black","orange","pink","grey"])
fig.show()
```

# **11. plot a graph that provides statistics on the intrest of customers wrt to product category**

```python
category_counts = diwali['Product_Category'].value_counts().reset_index()
category_counts.columns = ['Product_Category', 'Count']

fig = px.histogram(category_counts, x='Product_Category', y='Count', title='Interest of Customers by Product Category',color='Product_Category')
fig.show()
```

# **12. I would like to know the budget spent on each product category in descending order**

```python
category_spend = diwali.groupby('Product_Category')['Amount'].sum().reset_index().sort_values(by='Amount', ascending=False)

fig = px.histogram(category_spend, x='Product_Category', y='Amount', title='Budget Spent on Each Product Category (Descending)',color='Product_Category')
fig.show()
```
